```python
# Import libraries
import glob
import pandas as pd
```

```python
# Create a list of csv files list from a folder (called data)
path = './data'
csv_files = glob.glob(path + "/*.csv")

print(csv_files)
```

```python
# Read each CSV file into DataFrame
# This creates a list of dataframes
df_list = (pd.read_csv(file) for file in csv_files)

# Concatenate all DataFrames into a single DataFrame using pandas concat function
big_citibike_df   = pd.concat(df_list, ignore_index=True)
```

```python
# view big_df containing all citibike transactions for 2019
big_citibike_df
```

```python
# use pandas sample to sample a fraction of the big_citibike_df (20%)
small_citibike_df = big_citibike_df.sample(frac=.25)

small_citibike_df.info()
```

```python
# save small_df to csv file
small_citibike_df.to_csv('./sample_data/citi_bike_2019_frac_25.csv', index=False)
```
